```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
TP_loss = np.load('TP_loss.npy')
FP_loss = np.load('FP_loss.npy')
TN_loss = np.load('TN_loss.npy')
FN_loss = np.load('FN_loss.npy')
```

```python
def group(loss_array):
    x=[]
    y=[]
    y_mean = [0]*10
    for line in loss_array:
        for i, item in enumerate(line):
            x.append(i)
            y.append(item)

    for i, item in enumerate(y):
        y_mean[i%10] += item
    for i, _ in enumerate(y_mean):
        y_mean[i] = y_mean[i]/len(loss_array)
    return x, y, y_mean
```

```python
x1, y1, mean_1 = group(TP_loss)
x2, y2, mean_2 = group(FP_loss)
x3, y3, mean_3 = group(TN_loss)
x4, y4, mean_4 = group(FN_loss)
```
